```python
import requests
from bs4 import BeautifulSoup

# Define keywords for classification
ecommerce_keywords = ['cart', 'basket', 'checkout', 'product', 'shop', 'buy', 'ecommerce']
marketing_ops_keywords = ['campaign', 'marketing', 'seo', 'lead', 'ads', 'content', 'brand']

def classify_website(url):
    try:
        response = requests.get(url, timeout=5)
        content = response.text.lower()
        
        # Keyword matching
        ecom_score = sum(kw in content for kw in ecommerce_keywords)
        markops_score = sum(kw in content for kw in marketing_ops_keywords)
        
        if ecom_score > markops_score and ecom_score > 0:
            return 'e-commerce'
        elif markops_score > ecom_score and markops_score > 0:
            return 'marketing-ops'
        else:
            return 'unknown'
    except:
        return 'unreachable'
```

```python
df['Category'] = df['Website'].apply(classify_website)
```
